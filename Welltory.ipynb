```python
import pandas as pd
import openai
import os

# Установка ключа API OpenAI
openai.api_key = "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"


# Определение функции для анализа тональности отзыва
def analyze_review_sentiment(text):
    response = openai.Completion.create(
        engine="text-davinci-002",
        prompt=f"Please rate the following review on a scale of 1-100, with 1 being the most negative and 100 being the most positive(only one number without text \n\n{text}\n\nRating:",
        temperature=0.5,
        max_tokens=1,
        n=1,
        stop=None,
    )
    rating = response.choices[0].text.strip()
    return rating
pathname = "C:/Users/Vladislav/reviews/Welltory Test_Python Developer_App Reviews - Data"
# Определение имени входного файла и выходного файла
input_file = pathname + '.csv'
output_file = pathname +"_analyzed.csv"

# Чтение данных из входного файла
data = pd.read_csv(input_file,sep = ',')

# Применение функции для анализа тональности к столбцу с отзывами
data["rate"] = data["review text"].apply(lambda x: analyze_review_sentiment(str(x)))


# Сортировка данных по столбцу с оценкой в обратном порядке
data = data.sort_values("rate", ascending=False)

data.drop(['date','review text'], axis=1, inplace=True)
# Запись результатов в новый файл
data.to_csv(output_file, index=False)


# Вывод сообщения о завершении программы
#print(f"Результаты сохранены в файл {output_file}")
```
